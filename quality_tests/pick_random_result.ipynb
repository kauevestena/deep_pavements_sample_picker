```python
rootpath = '../../data/samples'

import os,random
from IPython.display import Image
```

```python
# put None if you want to pick a random category
category = 'road'
```

```python
cities = os.listdir(rootpath)

random_city = random.choice(cities)

samples = os.listdir(os.path.join(rootpath,random_city))

random_sample = random.choice(samples)

image = os.path.join(rootpath,random_city,random_sample,random_sample+'.png')

# show image
Image(image)
```

```python
categories = os.listdir(os.path.join(rootpath,random_city,random_sample,'clipped_detections'))

if not category:
    category = random.choice(categories)

imagelist = os.listdir(os.path.join(rootpath,random_city,random_sample,'clipped_detections',category))
    
random_image = random.choice(imagelist)

imagepath = os.path.join(rootpath,random_city,random_sample,'clipped_detections',category,random_image)

# show image
Image(imagepath)
```
